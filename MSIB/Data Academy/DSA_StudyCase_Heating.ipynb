## Linier Regresi

```python
#menentukan input variable

x=df[['Insulation', 'Temperature', 'Num_Occupants', 'Avg_Age', 'Home_Size']]

#menentukan target variable
y=df['Heating_Oil']
```

```python
#untuk split data kita menjadi training dan testing
from sklearn.model_selection import train_test_split
#untuk linier model
from sklearn.linear_model import LinearRegression
# untuk evaluasi model
from sklearn import metrics
```

```python
#spilt data kita menjadi training dan testing
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.3, random_state=0)
```

```python
x_train
```

```python
regressor=LinearRegression()

#train model
model_reg = regressor.fit(x_train, y_train)
```

```python
# Memprediksi dengan data test
y_pred_reg = model_reg.predict(x_test)
```

```python
#evaluasi linier regression
print('r2_score :', metrics.r2_score(y_test, y_pred_reg))
```

```python
#membuat script MAPE
def mean_absolute_percentage_error(y_true, y_pred):
  y_true, y_pred = np.array(y_true), np.array(y_pred)
  return np.mean(np.abs((y_true-y_pred)/y_true))*100
```

```python
print('MAPE :', mean_absolute_percentage_error(y_test, y_pred_reg))
```

## Hasil evaluasi pada model Linier regresion

Baik

R2 menunjukkan nilai 0.81 mendekati 1, artinya untuk menjadi exellent hanya kurang 0.2
Mape = 7.8. artinya tingkat error = 7.8% dan tingkat kesuksesan 92.2%

## pemodelan dengan MLP Regresor (ANN = artifisial Neural Network)

```python
from sklearn.neural_network import MLPRegressor
```

```python
#objek MLP Regresor
mlp_reg=MLPRegressor()

#training model
model_mlp=mlp_reg.fit(x_train,y_train)

#predict model
y_pred_mlp=model_mlp.predict(x_test)
```

```python
#evaluasi MLP regresor
print('r2_score :', metrics.r2_score(y_test, y_pred_mlp))
```

```python
print('MAPE :', mean_absolute_percentage_error(y_test, y_pred_mlp))
```
